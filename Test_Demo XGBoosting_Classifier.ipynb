# Test demo. for reproduced XGBoosting classifier

```python
from sklearn.datasets import load_iris  
from XGBM import XGboostingClassifier  
from utils import plot_decision_boundary  
```

1. Load Iris dataset from scikit-learn

```python
iris = load_iris()

data = iris.data[50:, 2:4]
label = iris.target[:100] 
```

2. Fit the model and show how the decision boundary changes over iterations

```python
for i in range(1,8):
    model = XGboostingClassifier(n_estimator = i, max_depth = 4, min_sample = 20, 
                                 num_bins = 20, min_score_gain = 0, gamma = 0, _lambda = 0, lr = 0.9)
    
    model.fit(data, label)
    plot_decision_boundary(model, data, label)
```
